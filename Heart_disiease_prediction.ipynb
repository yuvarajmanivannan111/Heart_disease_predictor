Implement Linear Reggression algorithm

Seggregating the features and the target

```python

Feature=Heart_df[['age','trestbps','chol','thalach','oldpeak','thal']]
Target=Heart_df['target']
Check_cor=Heart_df[['age','trestbps','chol','thalach','oldpeak','thal','target']]
```

Spliting the Data for training and testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(Feature,Target,test_size=0.2,random_state=42)
```

In this particular dataset scaling is required because the is a variation in digits between colums only for features

```python
from sklearn.preprocessing import StandardScaler
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

Training the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train_scaled,y_train)

```

Testing the Trained model

```python
y_pred = model.predict(X_test_scaled)
print(y_pred)
```

```python
# Display a few predictions
predictions = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(predictions.head())
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
pd.crosstab(y_test,y_pred)
```

Checking which Threshold performs better and giving higher accuracy

```python

def predict_threshold (model,X_test,thresholds):
   import numpy as np
   return np.where(model.predict_proba(X_test)[:,1]>thresholds,1,0)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
for thr in np.arange(0,1.0,0.1):
   y_predict = predict_threshold(model,X_test_scaled,thr)
   print("Threshold :",thr)
   print(confusion_matrix(y_test,y_predict))
   print(accuracy_score(y_test,y_predict))
```

For achieving a better or good accuracy the perfect threshold is 0.4 because it giving a best accuracy of 71%

```python
import numpy as np
y_predict_th4 = np.where(model.predict_proba(X_test_scaled)[:,1]>0.4,1,0) # Threshold is 0.3
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict_th4))
pd.crosstab(y_test,y_predict_th4)
```

Conclusion:
The default thershold 0.4 is giving the better accuracy so no need for changing the threshold this model giving 71.21% accuracy for 0.4 Threshold
